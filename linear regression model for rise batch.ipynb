```python
# here now we will start to build model , for this we have to define input features and output feature (target variable)

# when there are more than one feature is available as input we go for taking all features as input
X=df[["YearsExperience"]]
# x = df.drop(columns=["Salary"],axis=1)
```

```python
y=df.Salary
```

```python
#calling the linear regression model 
from sklearn.linear_model import LinearRegression
```

```python
# creating object of model 
model=LinearRegression()
model
```

```python
# after calling the model we are going to split the dataset for training ad testing 
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42) # splitting the data 
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# after splitting the data into training and testing , now we are going to train the model on training data
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
model.predict([[6.3]])
```

```python
# after training and testing the next step is to evaluate the model= evaluate model
m=model.coef_
c=model.intercept_
```

```python
# Compare with actual salaries (y_test)
print("Actual salaries:", y_test)
print("Predicted salaries:", y_pred)

# Or create a comparison table
import pandas as pd
results = pd.DataFrame({
    
    'Actual': y_test,
    'Predicted': y_pred,
    'Difference': y_test - y_pred
})

print(results)
```

```python
accuracy_of_model=model.score(X,y)*100
```

```python
t# as we know that the model try to fit best fit line , so here we will create a plot which will show best fit line
# plt.figure(figsize=(12, 6))

# plt.subplot(1, 2, 1)
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_train, model.predict(X_train), color='blue', linewidth=2, label='Regression line')
plt.title('Training Data - Linear Regression')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
mse=mean_squared_error(y_test, y_pred)
print("mean sq error:", mse)
r2=r2_score(y_test, y_pred)
print("r2 score is:", r2)
rmse = np.sqrt(mse)
print("root mse is:", rmse)
```
