```python
## HOW TO FIND MOST IMPORTANT FEATURES FOR THE MODEL : FEATURE IMPORTANCE  ##
```

```python
## We’ll use a tiny dataset predicting whether a person buys a car.
```

```python
from sklearn.ensemble import RandomForestClassifier  ## this is classification model like logistic reg, you can use logistic reg model also
data = {
    "Age":     [22, 25, 47, 52, 46, 56, 28, 32],
    "Salary":  [20000, 30000, 50000, 45000, 41000, 60000, 35000, 38000],
    "City":    [0, 1, 1, 0, 0, 1, 0, 1],
    "Purchased": [0, 0, 1, 1, 1, 1, 0, 0]
}
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
## first of all i will divide the features into input and output
X = df[["Age", "Salary", "City"]]
y = df["Purchased"]  ## dependant feature
```

```python
X  ## these are my input features 
```

```python
## now i will train the model on the input and output data (labelled data)
model = RandomForestClassifier(random_state=42)
model.fit(X, y)
```

```python
## now to calculate feature importance there is method
## model.feature_importance_
```

```python
feature_importances=model.feature_importances_
```

```python
feature_importances
```

```python
# as we can see the values but we are not exactly getting that which value represent which feature 
```

```python
## so lets create dataframe using pandas
feature_importances1 = pd.DataFrame({
    "Feature": X.columns,
    "Importance": model.feature_importances_
})
```

```python
feature_importances1   ## the magnitude of values represent how much the feature is important 
```
