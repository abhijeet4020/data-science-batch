```python
# ##| Feature          | Label Encoding              | One Hot Encoding                 |
# | ---------------- | --------------------------- | -------------------------------- |
#             When the categories have order or ranking         When categories are nominal (no order)
                                                                    #(ex. color, city, gender, country)
                  #(ex., Small < Medium < Large)

# | Output           | Single column with integers | Multiple columns with 0/1        |
#
# | Data Type        | Ordinal or Nominal          | Nominal only                     |
# | Order Introduced | Yes (0 < 1 < 2)             | No (pure binary)                 |
# | Good for         | Tree models(DT,RF)          | Linear, distance & neural models |
# | Issue            | False ranking               | More features created            |
```

```python
## one hot encoding create sparse matrix of binary numbers 0 and 1 , and label encoding create single columns with integers
## label encoding gives ranking like 1 ,2 3, 45,67, due to which the ML model may get misunderstanding about the importnace of feature
# like suppose there are 20 colors, if color red is coded as 0 and yellow is coded as 20 
#then machine may give more importance to yellow color than red 
```
