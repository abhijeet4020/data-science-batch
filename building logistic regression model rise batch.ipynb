```python
#as we can see the Cabin colun contain 687 nul values so what we will do is fill 
# those null values with most repeated cabin class for which we use "mode"
df['Cabin'].fillna(df['Cabin'].mode() [0], inplace=True)
```

```python
df.Cabin.isnull().sum()   #now we can see no null value is present into cabin column
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df['Embarked'].isnull().sum() #now we can see no null value is present
```

```python
#why we use [0] , when we go for 
df.Embarked.mode()
```

```python
#see above it return something like this , where s is located at index 0,
# so [0] â†’ This selects the first value from the Series returned by .mode()
```

```python
# now fill the null values in age column as mean of all ages
df["Age"].fillna(df["Age"].median(), inplace=True)
```

```python
df.Age.isna().sum() #now we can see no null value is present
```

```python
                          # FEATURE SELECTION AND FEATURE ENGINEERING

#after filling all null values we have to go for now selecting required features , feature engineering
#which include converting the categorical data into numerical data
```

```python
# select the features which are really affecting the models output feature 
#PassengerId	Survived	Pclass	Name	Sex	Age	SibSp	Parch	Ticket	Fare	Cabin	Embarked
newdf=df[["Survived","Pclass","Sex","Age","SibSp", "Parch", "Fare","Embarked"]]
# this is feature selection , selcting the most important features 
```

```python
newdf.head() # at this new dataset few features are categorical so we have to convert them into numerical
```

```python
newdf.Embarked.unique()
```

```python
newdf['Sex'] = newdf['Sex'].map({'male': 1, 'female': 0})
```

```python
newdf['Sex']
```

```python
newdf['Embarked'] = newdf['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
newdf['Embarked']
```

```python
newdf.head()
```

```python
#now we can see all the columns are in numerical 
```

```python
# train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = newdf.drop('Survived', axis=1)
```

```python
y = newdf['Survived']
#you can also write y=newdf.Survived
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,        # 20% test data
    random_state=42,      # keeps results same every time
)
```

```python
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
#import model and train it 
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train) # this is training model
```

```python
y_pred = model.predict(X_test) #this is testing the model 
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Accuracy
acc = accuracy_score(y_test, y_pred)
print("\nAccuracy:", acc)

# Classification Report
print("\nClassification Report:\n", classification_report(y_test, y_pred))

```
